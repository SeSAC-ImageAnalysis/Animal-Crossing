```python
# 라이브러리 불러오기
import torch
from PIL import Image

# gpu 사용 가능시에 cuda 선택
device = "cuda" if torch.cuda.is_available() else "cpu"
print(device)

```

```python
%pwd
```

```python
# 모델, 로라 불러오기

model_path = "./model/anythingelseV4_v45.safetensors"
lora_path = './sd_lora/last_of_last-10.safetensors'

from diffusers import DiffusionPipeline, StableDiffusionImg2ImgPipeline, AutoPipelineForImage2Image

pipeline = StableDiffusionImg2ImgPipeline.from_single_file(model_path, use_safetensors=True, variant="fp32", torch_dtype=torch.float16,).to(device)
pipeline.load_lora_weights(lora_path)
```

```python
# 에러나서 막아놓음
#pipeline.enable_xformers_memory_efficient_attention()
```

```python
init_image = Image.open("./img/남자1.png").convert("RGB")
init_image = init_image.resize((768, 512))
```

```python
pipeline.safety_checker = None
pipeline.requires_safety_checker = False
```

```python
positive_prompt = "masterpiece, best quality, high resolution illustration, high quality, 8k, 1male, black hair, short cut, upper body, pov, portrait, looking at viewer, smile, front view, from front, black eyes, white background"
negative_prompt = "(worst quality, low quality:1.3), (nsfw),  look back, hands, odd eye, monocfrome, gray scale, lowres, bad anatomy, bad hands, text, error, missing fingers, extra digit, fewer digits, cropped, worst quality, jpeg artifacts, signature, watermark, username, liquid body, liquid tongue, disfigured, malformed, mutated, anatomical nonsense, text font ui, long neck, blurred, lowers, bad proportions, bad shadow, uncoordinated body, unnatural body, fused hand, disappearing arms, red background, flowers"
```

```python
# 10장 생성

# result = pipeline(prompt = "horse", generator=torch.manual_seed(0)).images
# , generator=torch.manual_seed(3832868744)


result = pipeline(prompt = positive_prompt, negative_prompt=negative_prompt, image=init_image,  strength=0.75, guidance_scale=5, num_inference_steps=70, height=512, width=512, num_images_per_prompt=10).images
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 5, figsize=(8,4))

for i in range(2):
    for j in range(5):
        axes[i, j].imshow(result[i*j+1])
        axes[i, j].axis('off')
        fig.tight_layout()
```

```python
def predict(img_path, positive_prompt, negative_prompt):
    init_image = Image.open(img_path).convert("RGB")
    init_image = init_image.resize((768, 512))

    result = pipeline(prompt = positive_prompt, negative_prompt=negative_prompt, image=init_image,  strength=0.75, guidance_scale=5, num_inference_steps=70, height=128, width=128, num_images_per_prompt=1).images
    return result[0]
```
